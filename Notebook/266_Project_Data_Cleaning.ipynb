```python
import numpy as np
import os
folder_path = '/content/drive/My Drive/266_project/data/Electronics_5.json.gz'
```

```python

import gzip
import pandas as pd
import json

def parse(path):
  with gzip.open(path, 'rt', encoding='utf-8') as g:
    for l in g:
      yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    if i % 1000000 == 0:
      print(i)
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

```

```python
df = getDF(folder_path)
```
